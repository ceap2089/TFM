```python
import joblib as jl
import numpy as np
import pandas as pd
```

```python
x_out_train_a = jl.load('C:/Users/TSE LATAM/Documents/Navarra/TFM/notebooks_3/dataset_2909.pkl')
x_out_train_b = jl.load('C:/Users/TSE LATAM/Documents/Navarra/TFM/notebooks_3/dataset_0311.pkl')
x_out_test = jl.load('C:/Users/TSE LATAM/Documents/Navarra/TFM/notebooks_3/dataset_1812.pkl')
```

```python
# Separar los datos en arrays de numpy
images_train_a = x_out_train_a['images']
intervals_train_a = x_out_train_a['intervals']
labels_train_a = x_out_train_a['labels']

images_train_b = x_out_train_b['images']
intervals_train_b = x_out_train_b['intervals']
labels_train_b = x_out_train_b['labels']

images_test = x_out_test['images']
intervals_test = x_out_test['intervals']
labels_test = x_out_test['labels']
```

```python
# Concatenar los arrays
images_train = np.concatenate((images_train_a, images_train_b), axis=0)
intervals_train = np.concatenate((intervals_train_a, intervals_train_b), axis=0)
labels_train = np.concatenate((labels_train_a, labels_train_b), axis=0)
```

```python
# Normalizar los datos de entrenamiento y prueba
images_train = images_train.astype('float32') / 255.0
images_test = images_test.astype('float32') / 255.0
```

```python
import tensorflow as tf
from functools import partial
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Suponiendo que los datos ya están cargados y concatenados
# images_train, images_test, labels_train, labels_test deben estar definidos previamente

# Asegurarse de que las imágenes tienen la forma correcta
images_train = np.expand_dims(images_train, axis=-1)  # (4800, 1601, 64, 1)
images_test = np.expand_dims(images_test, axis=-1)  # (2400, 1601, 64, 1)

# Codificar las etiquetas
label_encoder = LabelEncoder()
labels_train_categorical = label_encoder.fit_transform(labels_train)
labels_test_categorical = label_encoder.transform(labels_test)

# Verificar la forma de las imágenes
print(f"Forma de images_train: {images_train.shape}")
print(f"Forma de images_test: {images_test.shape}")

DefaultConv2D = partial(
    tf.keras.layers.Conv2D,
    kernel_size=3,
    padding='same',
    activation='relu',
    kernel_initializer='he_normal'
)

model_classifier = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1601, 64, 1)),
    DefaultConv2D(filters=32),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=64),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=128),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=256),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(rate=0.5),
    tf.keras.layers.Dense(units=512, activation='relu'),
    tf.keras.layers.Dense(units=len(np.unique(labels_train_categorical)), activation='softmax')
])

model_classifier.summary()  # Verificar la forma de las salidas después de cada capa

callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', min_delta=0.01, patience= 20)

model_classifier.compile(loss='sparse_categorical_crossentropy',
                optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
                metrics=['accuracy'])

# Train the model
history = model_classifier.fit(images_train, labels_train_categorical, epochs=100, 
                    validation_data=(images_test, labels_test_categorical), callbacks=[callback])

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.show()

# Generate predictions and plot confusion matrix
y_pred = model_classifier.predict(images_test)
y_pred_classes = np.argmax(y_pred, axis=1)

cm = confusion_matrix(labels_test_categorical, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.title('Confusion Matrix')
plt.show()

# Calcula y muestra la exactitud y la tasa de errores
accuracy = np.trace(cm) / np.sum(cm)
print(f"Exactitud del modelo: {accuracy * 100:.2f}%")
error_rate = 1 - accuracy
print(f"Tasa de errores: {error_rate * 100:.2f}%")

```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, BatchNormalization, LeakyReLU
from tensorflow.keras.models import Model

def build_light_autoencoder():
    # Define la entrada con el nuevo tamaño de imagen
    input_img = Input(shape=(1600, 64, 1))

    # Encoder
    x = Conv2D(16, (3, 3), padding='same')(input_img)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(32, (3, 3), padding='same')(x)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), padding='same')(x)
    x = Conv2D(64, (3, 3), padding='same')(x)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), padding='same')(x)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(32, (3, 3), padding='same')(x)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(16, (3, 3), padding='same')(x)  # Reduce number of filters
    x = LeakyReLU(alpha=0.1)(x)
    x = BatchNormalization()(x)
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = Model(input_img, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')

    return autoencoder

# Construir el modelo
autoencoder = build_light_autoencoder()
autoencoder.summary()
```

```python
import tensorflow as tf
from functools import partial
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Codificar las etiquetas
label_encoder = LabelEncoder()
labels_train_categorical = label_encoder.fit_transform(labels_train)
labels_test_categorical = label_encoder.transform(labels_test)

# labels_train_categorical = labels_train_categorical[indices_x_train_a_position]
# labels_test_categorical = labels_test_categorical[indices_y_test_a_position]

DefaultConv2D = partial(
    tf.keras.layers.Conv2D,
    kernel_size=3,
    padding='same',
    activation='relu',
    kernel_initializer='he_normal'
)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1600, 64, 1)),
    DefaultConv2D(filters=32),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=64),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=128),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    DefaultConv2D(filters=256),
    tf.keras.layers.MaxPool2D(pool_size=(2,2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(rate=0.5),
    tf.keras.layers.Dense(units=512, activation='relu'),
    tf.keras.layers.Dense(units=len(np.unique(labels_train_categorical)), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy',
                optimizer=tf.keras.optimizers.SGD(learning_rate=1e-3),
                metrics=['accuracy'])


# Train the model
history = model.fit(predictions_train, labels_train_categorical, epochs=100, 
                    validation_data=(predictions_test, labels_test_categorical), callbacks=[callback])

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.show()

# Generate predictions and plot confusion matrix
y_pred = model.predict(predictions_test)
y_pred_classes = np.argmax(y_pred, axis=1)

cm = confusion_matrix(labels_test_categorical, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.xticks(rotation=90)  # Rotate x-axis labels by 90 degrees
plt.title('Confusion Matrix')
plt.show()

# Calcula y muestra la exactitud y la tasa de errores
accuracy = np.trace(cm) / np.sum(cm)
print(f"Exactitud del modelo: {accuracy * 100:.2f}%")
error_rate = 1 - accuracy
print(f"Tasa de errores: {error_rate * 100:.2f}%")
```
